```python
bracket_matching = {
    "[": "]",
    "{": "}",
    "<": ">",
    "(": ")",
}
bracket_matching_reversed = {v: k for k, v in bracket_matching.items()}
error_score = {
    "(": 3,
    "[": 57,
    "{": 1197,
    "<": 25137
}

completion_scores = {
    "(": 1,
    "[": 2,
    "{": 3,
    "<": 4,
}
error_score_sum = 0
incomplete_scores = []
for line in input_text.split("\n"):
    blocks = []
    print(line)
    valid = True
    for n, char in enumerate(line):
        if char in bracket_matching.keys():
            blocks.append(n)
        elif char in bracket_matching.values():
            idx = blocks.pop()
            opening_char = line[idx]
            if char != bracket_matching[opening_char]:
                error_score_sum += error_score[bracket_matching_reversed[char]]

                print(f"Incorrect character found! {char} instead of {bracket_matching[opening_char]}")
                # print(line)
                print("~" * n + "^")
                print(blocks)
                valid = False
                break
            else:
                # print("Block complete")
                # print(line)
                print(" "* idx + opening_char + "-" * ((n-idx)-1) + bracket_matching[opening_char])
        else:
            raise ValueError(char)

    if blocks and valid:
        incomplete_score = 0
        print(f"{line=}")
        print(f"{blocks=}")
        while blocks and (last_opener := blocks.pop()) is not None:
            print(f"{last_opener=} {line[last_opener]=}")
            incomplete_score *= 5
            incomplete_score += completion_scores[line[last_opener]]
            # print(f"Incomeplete: {incomplete_score}")
        incomplete_scores.append(incomplete_score)
        print(f"Completed with incomplete score of {incomplete_score}")


error_score_sum
```
