```python
from pathlib import Path

input_no = "05"
input_file = Path(f"../input_data/{input_no}")
```

```python
lines = []
# for input_line in """0,9 -> 5,9
# 8,0 -> 0,8
# 9,4 -> 3,4
# 2,2 -> 2,1
# 7,0 -> 7,4
# 6,4 -> 2,0
# 0,9 -> 2,9
# 3,4 -> 1,4
# 0,0 -> 8,8
# 5,5 -> 8,2""".split("\n"):


for input_line in input_file.read_text().split("\n"):
    a, b = input_line.split(" -> ")
    ax, ay = [int(x) for x in a.split(',')]
    bx, by = [int(x) for x in b.split(',')]


    line = ((ax, ay), (bx, by))
    lines.append(line)
```
