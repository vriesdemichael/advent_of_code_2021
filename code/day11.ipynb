```python
def illuminate_adjacent(x, y, mat, skip_list):
    adjacent_list = [(ax, ay) for ax, ay in [
        (x-1, y-1),
        (x-1, y+0),
        (x-1, y+1),
        (x+0, y-1),
        # (x+0, y+0),
        (x+0, y+1),
        (x+1, y-1),
        (x+1, y+0),
        (x+1, y+1),
    ] if 0 <= ax < 10 and 0 <= ay < 10  # Only in range of the matrix
        and (ax, ay) not in skip_list]  # Only when the octopus did not illuminate in this round yet
    for ax, ay in adjacent_list:
        mat[ax, ay] += 1

    mat[x, y] = 0


total_illuminations = 0
print("Starting")
for _ in tqdm(range(100)):
    print(f"Round {_}")
    skip_list = set()

    # Do the +1
    for x in range(10):
        for y in range(10):
            octopus_matrix[x, y] += 1

    while octopus_matrix.max() > 9:
        for x in range(10):
            for y in range(10):
                if octopus_matrix[x, y] > 9:
                    skip_list.add((x, y))
                    total_illuminations += 1
                    illuminate_adjacent(x, y, octopus_matrix, skip_list)

```
